```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load my dataset
df = pd.read_csv("/Users/mac/Downloads/crimebystatecombinedwithunemployment.csv")
```

```python
# Describe dataset
df.describe()
```

```python

# Calculate the correlation coefficient between unemployment and crime rate
correlation = df["unemployment"].corr(df["violent total"])
print("The correlation between unemployment and violent total:", correlation)


```

```python
# Create scatter graph
plt.style.use("bmh")
plt.figure(figsize=(8, 6))
plt.scatter(df["unemployment"], df["violent total"])
plt.xlabel("Unemployment")
plt.ylabel("Violent Total")
plt.title("Scatter Plot of Unemployment vs Total Crime/Violence")
plt.show()
```

```python
# Create histogram
plt.hist(df["unemployment"], bins=10, alpha=0.7, label="Unemployment", color="blue")
plt.hist(df["violent total"], bins=10, alpha=0.7, label="Violence", color='red')
plt.xlabel("Unemployment")
plt.ylabel("Violent Total")
plt.title("Histogram of Unemployment vs Total Crime/Violence")
plt.legend()
plt.show()
```

```python
# Create heatmap
sns.heatmap(df[["unemployment", "violent total"]], annot=True, cmap= "coolwarm" )
plt.title("Heat Map between Unemployment and Total Crime/Violence")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Total Violent")
plt.show()

```
